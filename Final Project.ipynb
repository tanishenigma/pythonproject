```python
import numpy as np
import pandas as pd
import os
from PIL import Image
import tensorflow as tf
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPool2D, Flatten, Dense, Dropout
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, classification_report

base_path = './gtsrb-german-traffic-sign'
train_csv_path = os.path.join(base_path, 'Train.csv')
test_csv_path = os.path.join(base_path, 'Test.csv')
train_data_path = os.path.join(base_path, 'Train')
test_data_path = os.path.join(base_path, 'Test')

print("Loading Training Data CSV...")
train_df = pd.read_csv(train_csv_path)
print("Training Data CSV Loaded.")

if 'Path' in train_df.columns:
    train_df['Path'] = train_df['Path'].apply(lambda x: x.replace('Train/', ''))

data = []
labels = []

print("Loading training images...")
for index, row in train_df.iterrows():
    img_path = os.path.join(train_data_path, row['Path'])
    if not os.path.exists(img_path):
        print(f"Image path not found: {img_path}")
        continue
    try:
        image = Image.open(img_path).resize((30, 30))
        data.append(np.array(image))
        labels.append(row['ClassId'])
    except Exception as e:
        print(f"Error loading image {img_path}: {e}")

data = np.array(data)
labels = np.array(labels)

print(f"Data shape: {data.shape}, Labels shape: {labels.shape}")

X_train, X_val, y_train, y_val = train_test_split(data, labels, test_size=0.2, random_state=42)
print(f"Training set shape: {X_train.shape}, Validation set shape: {X_val.shape}")

y_train = to_categorical(y_train, num_classes=43)
y_val = to_categorical(y_val, num_classes=43)

model = Sequential([
    Conv2D(filters=32, kernel_size=(5, 5), activation='relu', input_shape=X_train.shape[1:]),
    Conv2D(filters=64, kernel_size=(5, 5), activation='relu'),
    MaxPool2D(pool_size=(2, 2)),
    Dropout(rate=0.15),
    Conv2D(filters=128, kernel_size=(3, 3), activation='relu'),
    Conv2D(filters=256, kernel_size=(3, 3), activation='relu'),
    MaxPool2D(pool_size=(2, 2)),
    Dropout(rate=0.20),
    Flatten(),
    Dense(512, activation='relu'),
    Dropout(rate=0.25),
    Dense(43, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

model.summary()

print("Training the model...")
history = model.fit(X_train, y_train, batch_size=64, epochs=20, validation_data=(X_val, y_val))

def plot_performance(history):
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Model Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Model Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid()

    plt.show()

plot_performance(history)

print("Loading Test Data CSV...")
test_df = pd.read_csv(test_csv_path)
print("Test Data CSV Loaded.")

test_labels = test_df["ClassId"].values
test_imgs = test_df["Path"].values

X_test = []
print("Loading test images...")
for img in test_imgs:
    img_path = os.path.join(base_path, img)
    if not os.path.exists(img_path):
        print(f"Test image not found: {img_path}")
        continue
    try:
        image = Image.open(img_path).resize((30, 30))
        X_test.append(np.array(image))
    except Exception as e:
        print(f"Error loading test image {img_path}: {e}")

X_test = np.array(X_test)   

predictions = np.argmax(model.predict(X_test), axis=-1)

accuracy = accuracy_score(test_labels, predictions)
print(f"Test Accuracy: {accuracy:.2f}")
print(classification_report(test_labels, predictions))

model.save("traffic_sign_classifier.h5")
print("Model saved as traffic_sign_classifier.h5")
```

```python
import os
import numpy as np
import ttkbootstrap as ttk  # For modern UI
from ttkbootstrap.constants import *
from tkinter import filedialog, messagebox
from PIL import Image, ImageTk, ImageFilter, ImageOps
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image

# Define class names
class_names = [
    'Speed Limit (20 km/h)', 'Speed Limit (30 km/h)', 'Speed Limit (50 km/h)',
    'Speed Limit (60 km/h)', 'Speed Limit (70 km/h)', 'Speed Limit (80 km/h)',
    'End of Speed Limit (80 km/h)', 'Speed Limit (100 km/h)', 'Speed Limit (120 km/h)',
    'No Overtaking (Two Cars)', 'No Overtaking for Trucks and Cars', 'Right of the Way',
    'Priority Road', 'Yield Right of Way', 'Stop', 'Prohibited for All Vehicles',
    'Tractors and Trucks > 3.5 Tons Prohibited', 'Entry Prohibited', 'Danger',
    'Single Curve Left', 'Single Curve Right', 'Double Curve', 'Rough Road',
    'Slippery Road', 'Road Narrows', 'Construction Site', 'Signal Lights Ahead',
    'Pedestrian Walk', 'Children', 'Bicycle Crossing', 'Snowfall',
    'Wild Animal Crossing', 'Restricted: No Stopping',
    'Mandatory Direction of Travel (Right)', 'Mandatory Direction of Travel (Left)',
    'Mandatory Direction of Travel (Front)', 'Mandatory Direction of Travel (Front-Right)',
    'Mandatory Direction of Travel (Front-Left)', 'Mandatory Direction of Travel (Down-Right)',
    'Mandatory Direction of Travel (Down-Left)', 'Traffic Circle',
    'End of No Passing Zone', 'End of No Passing Zone for Trucks'
]

try:
    model = load_model('/home/feather/Documents/Python Project/traffic_sign_classifier.h5')
    print("Model loaded successfully!")
except Exception as e:
    print(f"Error loading model: {e}")
    raise

def preprocess_image(img_path):
    try:
        img = image.load_img(img_path, target_size=(30, 30)) 
        img_array = image.img_to_array(img)
        img_array = np.expand_dims(img_array, axis=0) 
        img_array = img_array / 255.0 
        return img_array
    except Exception as e:
        raise ValueError(f"Error preprocessing image: {e}")

def predict_traffic_sign(img_path):
    try:
        img_array = preprocess_image(img_path)

        img = Image.open(img_path)
        img = img.resize((300, 300))

        blurred_img = img.filter(ImageFilter.GaussianBlur(5))
        img_tk = ImageTk.PhotoImage(blurred_img)
        img_label.config(image=img_tk)
        img_label.image = img_tk

        prediction = model.predict(img_array)
        predicted_class = np.argmax(prediction)
        confidence = np.max(prediction)

        prediction_text = f"Predicted: {class_names[predicted_class]} (Confidence: {confidence:.2f})"
        result_label.config(text=prediction_text)

        if confidence < 0.8:
            messagebox.showwarning("Uncertainty", "Prediction confidence is low. Please review manually.")

        # Plot confidence scores
        plt.figure(figsize=(15, 6))
        plt.bar(range(len(class_names)), prediction[0], color="red")
        plt.xticks(range(len(class_names)), class_names, rotation=90, fontsize=8)
        plt.ylabel("Confidence", color="black")
        plt.title("Prediction Confidence for Each Class", color="black")
        plt.tight_layout()
        plt.show()

    except Exception as e:
        messagebox.showerror("Error", f"An error occurred during prediction: {e}")

def select_sample_image():
    sample_dir = "./sample_images"  
    if not os.path.exists(sample_dir):
        os.makedirs(sample_dir) 

    img_path = filedialog.askopenfilename(
        initialdir=sample_dir,
        title="Select Sample Image",
        filetypes=[("Image Files", "*.png *.jpg *.jpeg")]
    )
    if img_path:
        predict_traffic_sign(img_path)

root = ttk.Window(themename="darkly") 
root.title("Traffic Sign Classifier")
root.geometry("900x700")

bg_color = "#1c1c1c"  # Dark background
text_color = "#f44336"  # Bright red for text
button_color = "#1c1c1c"  # Dark gray for buttons

# Apply background color
root.configure(bg=bg_color)

# Heading
title_label = ttk.Label(
    root,
    text="Traffic Sign Classifier",
    font=("Helvetica", 24, "bold"),
    anchor="center",
    foreground=text_color,
    background=bg_color
)
title_label.pack(pady=20)

# Buttons and labels
button_frame = ttk.Frame(root, padding=20, style="primary.TFrame")
button_frame.pack(pady=10)

sample_button = ttk.Button(
    button_frame, 
    text="Select Sample Image", 
    command=select_sample_image, 
    bootstyle="danger"  # Red Button
)
sample_button.grid(row=0, column=0, padx=10)

img_label = ttk.Label(root, text="Upload a sample image to display here.", anchor="center", background=bg_color, foreground=text_color)
img_label.pack(pady=20)

result_label = ttk.Label(root, text="Prediction result will appear here.", font=("Helvetica", 14), background=bg_color, foreground=text_color)
result_label.pack(pady=20)

# Run the application
root.mainloop()
```
